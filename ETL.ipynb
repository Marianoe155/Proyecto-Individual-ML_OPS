```python
import nltk # importo la libreria "nltk" para realizar el analisis de sentimiento 
nltk.download("vader_lexicon")
from nltk.sentiment.vader import SentimentIntensityAnalyzer # importo la funcion 
```

```python
# guardo la funcion que necesito en una variable 
sia = SentimentIntensityAnalyzer() 
```

```python
# Desarrollo una funcion para asignar nuestra escala de analisis de sentimiento 
def get_sentiment_score(text):
    if pd.isnull(text) or text == '':
        return 1  # devuelve 1 si encontramos vacios o Nan
    elif isinstance(text, str):
        sentiment = sia.polarity_scores(text)
        compound_score = sentiment["compound"]
        if compound_score >= 0.05:
            return 2  # devolvemos 2 si considero que es positiva 
        elif compound_score <= -0.05:
            return 0  # devolvemos 0 si considero que es negativa 
        else:
            return 1
    else:
        return 1  # devuelve 1 si consideramos que es neutral
```

```python
# la columna review necesito pasarla a string
df_1["review"] = df_1["review"].astype(str)
```

```python
# aplico la funcion a la columna "review" para crear una nueva columna "sentiment_analysis" con los score correspondiente
df_1["sentiment_analysis"] = df_1["review"].apply(get_sentiment_score)
```

```python
df_1.sentiment_analysis.value_counts() # observo si estan los 3 valores de nuestra escala 
```

```python
df_1.head(1) # veo que si se agrego la columna
```

```python
df_1["AÃ±o_posteo"].fillna(0, inplace= True) # donde tengo Nan en la columna "aÃ±o_postero", colocamos un "0"
```
